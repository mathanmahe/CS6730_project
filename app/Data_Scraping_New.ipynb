```python
import json

# Open and read the JSON file
with open('/content/drive/My Drive/Data visualization/top250.json', 'r') as file:
    data = json.load(file)

# Extract the list of IDs
movie_ids = [item['id'] for item in data['items']]


# Print the list of IDs
print(movie_ids)
```

```python
import requests

api_key = "k_c94i0nte"  # Replace with your actual API key
base_url = "https://imdb-api.com/{lang}/API/Title/{apiKey}/{id}/{options}"

# Assuming movie_ids is the list of IDs you extracted from the original JSON
detailed_movies = []

# url = base_url.format(lang="en", apiKey=api_key, id="tt0111161", options="")
# response = requests.get(url)
# if response.status_code == 200:
#     detailed_movies.append(response.json())

for movie_id in movie_ids:
    url = base_url.format(lang="en", apiKey=api_key, id=movie_id, options="")
    response = requests.get(url)
    if response.status_code == 200:
        detailed_movies.append(response.json())
    else:
        print(f"Failed to fetch data for ID {movie_id}")
```

```python
with open('/content/drive/My Drive/Data visualization/detailed_movies.json', 'w') as file:
    json.dump(detailed_movies, file, indent=4)
```

```python
# merge data from two files
movie_details_map = {item['id']: item for item in data['items']}


for movie in detailed_movies:
    movie_id = movie['id']

    if movie_id in movie_details_map:
        additional_details = movie_details_map[movie_id]

        # Merge additional details
        for key in ['rank', 'imDbRating', 'imDbRatingCount', 'title', 'fullTitle', 'year', 'image', 'crew']:
            movie[key] = additional_details.get(key)

with open('/content/drive/My Drive/Data visualization/detailed_movies.json', 'w') as file:
    json.dump(detailed_movies, file, indent=4)
```

```python
# query gender data
with open('/content/drive/My Drive/Data visualization/detailed_movies.json', 'r') as file:
    detailed_movies = json.load(file)

import requests

api_key = "k_c94i0nte"  # Replace with your actual API key
name_url = "https://imdb-api.com/{lang}/API/Name/{apiKey}/{id}"

for movie in detailed_movies:
    for actor in movie.get("actorList", []):
        actor_id = actor["id"]
        url = name_url.format(lang="en", apiKey=api_key, id=actor_id)
        response = requests.get(url)

        if response.status_code == 200:
            actor_data = response.json()
            role = actor_data.get("role", "").lower()
            if "actress" in role:
                actor["gender"] = "Female"
            elif "actor" in role:
                actor["gender"] = "Male"

with open('/content/drive/My Drive/Data visualization/detailed_movies_genders_updated.json', 'w') as file:
    json.dump(detailed_movies, file, indent=4)
```

```python
# find the bechdel api rating

from concurrent.futures import ThreadPoolExecutor
import requests

import json

# API request
def fetch_data(id):
    url = f"http://bechdeltest.com/api/v1/getMovieByImdbId?imdbid={id.replace('tt','')}"
    try:
        response = requests.get(url)
        response.raise_for_status()  # throw error on failure
        data = response.json()
        return data.get('rating', 'NA'), data.get('id', 'NA')
    except requests.RequestException as e:
        print(f"Error fetching data for ID: {id}, Error: {e}")
        return 'NA', 'NA'

# Parallel request for reducing time
with ThreadPoolExecutor(max_workers=10) as executor:
    results = list(executor.map(fetch_data, movie_ids))

# # extract lists
# ratings = [result[0] for result in results]
# bechdelIds = [result[1] for result in results]

with open('/content/drive/My Drive/Data visualization/detailed_movies_genders_updated.json', 'r') as file:
    detailed_movies = json.load(file)

for movie, (rating, bechdel_id) in zip(detailed_movies, results):
    movie['BechdelRating'] = rating
    movie['BechdelID'] = bechdel_id

with open('/content/drive/My Drive/Data visualization/detailed_movies_genders_bechdel_updated.json', 'w') as file:
    json.dump(detailed_movies, file, indent=4)
```

```python
import csv

# Read FemaleDirectors.csv and create a set of female director names
female_directors = set()
with open('/content/drive/My Drive/Data visualization/FemaleDirectors.csv', 'r') as file:
    csv_reader = csv.reader(file)
    next(csv_reader, None)  # Skip header row if there is one
    for row in csv_reader:
        female_directors.add(row[0].strip())  # Assuming name is in the first column

print(female_directors)
```
